```python
import pandas as pd
```

```python
Jan = pd.read_csv('Data/JC-201801-citibike-tripdata.csv')
Feb = pd.read_csv('Data/JC-201802-citibike-tripdata.csv')
Mar = pd.read_csv('Data/JC-201803-citibike-tripdata.csv')
Apr = pd.read_csv('Data/JC-201804-citibike-tripdata.csv')
May = pd.read_csv('Data/JC-201805-citibike-tripdata.csv')
Jun = pd.read_csv('Data/JC-201806-citibike-tripdata.csv')
Jul = pd.read_csv('Data/JC-201807-citibike-tripdata.csv')
Aug = pd.read_csv('Data/JC-201808-citibike-tripdata.csv')
Sep = pd.read_csv('Data/JC-201809-citibike-tripdata.csv')
Oct = pd.read_csv('Data/JC-201810-citibike-tripdata.csv')
Nov = pd.read_csv('Data/JC-201811-citibike-tripdata.csv')
Dec = pd.read_csv('Data/JC-201812-citibike-tripdata.csv')
# Apr.head()
```

```python
df = pd.concat([Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec])
# df.head()
```

```python
df.to_csv("tripdata.csv", index=False)
```
